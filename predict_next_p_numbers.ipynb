```python

import numpy as np

from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler

from numpy import transpose
from numpy.linalg import inv


pi = np.pi
T = np.transpose

# choose how many next numbers you want to predict
next_p = 10

y = np.array([0,1,2,3,2,5,8,11,8,17,26])
# y[i] = 3*y[i-4] + 2  is the sequence pattern above
# by the way, this y goes on like 35, 26, 53, 80, 107 ---
# i'll just systematically make generate_y function to re-make the sequence above

def generate_y(add_n_num):

    y = [0,1,2,3]

    for _ in range(add_n_num):
        t = len(y)
        next_num = 3*y[t-4] + 2 #- 2*y[t-3] + 2
        # make sure {max lag <= len(initial y)}
        y.append(next_num)

    return y

add_n_num = 7
y =generate_y(add_n_num)
#print(y)


#print(len(y))
print('y = ', y)

def generate_lag_data(y, k_lags):

    leng = len(y) - k_lags

    lag_mtx = np.empty([len(y)-k_lags, k_lags+1])

    for i in range(len(lag_mtx)):

        lag_mtx[i] = y[i    : i+ k_lags+1  ]

    return lag_mtx



k_lags = 5  # set a bit high, otherwise a model doesn't fit well
lag_mtx = generate_lag_data(y, k_lags)

#print(lag_mtx.shape)
#print()
#print(lag_mtx)
#print()

X = lag_mtx[:, 0:k_lags]
target = lag_mtx[:, -1]
#print(X.shape)
#print()
#print(target.shape)
#print()
#print(X)
#print()


###################################################

def find_intercept_coefs(X, y):

    ones_col = np.ones([len(X), 1])

    X = np.concatenate((ones_col, X), axis=1)

    Y = transpose([y])

    XT = transpose(X)

    XTX = np.dot(XT, X)

    invXTX = np.linalg.pinv(XTX)

    invXTXXT = np.dot(invXTX, XT)

    intercept_coefs = np.dot(invXTXXT, Y)

    return intercept_coefs



z = find_intercept_coefs(X, target)
intercept = z[0][0]
coefs = z[1:]

#print('intercept = ',intercept)
print()
#print('coefs = ',coefs)
print()


'''
inp = np.array([[8,11, 8,17,26],
               [11,8,17,26,35],
               [8,17,26,35,26],
               [17,26,35,26,53],
               [26,35,26,53,80],
               [35,26,53,80,107]])
'''


# we predict next number with the last 'k_lags' sequence as input
#inp = np.array([8,11,8,17,26])

# do it systematically
inp = y[-k_lags:]

pred_next_number = np.dot( inp, coefs  ) + intercept
#print('next number might be = ', pred_next_number)

## if y is non linear, is won't go well.

#print(inp)

################################################

def predict_next_p_num(next_p, y, k_lags, coefs, intercept):

    res = []

    predictor = y[-k_lags:]

    for i in range(next_p):
        next_num = (np.dot( predictor[i:], coefs ) + intercept)[0]
        res.append(next_num)
        predictor.append(next_num)

    return res


res = predict_next_p_num(next_p, y, k_lags, coefs, intercept)

txt = (f"predict next {next_p} numbers  " , res    )
print(txt)

print()
print()
print()

```
